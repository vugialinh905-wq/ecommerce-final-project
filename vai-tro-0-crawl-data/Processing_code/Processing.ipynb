## 4.4. Xử lý ngôn ngữ tự nhiên (NLP)

```python
def clean_text(text_in: str) -> str:
    """
    Clean & normalize text (Vietnamese-friendly).
    - Remove HTML entities
    - Remove mentions, hashtags, URLs
    - Remove emoji & special symbols
    - Normalize whitespace
    - Keep all words (no stopword / length filtering)
    """

    # 1. Chuyển đổi các ký tự HTML: &amp; → &, &lt; → <
    text_in = html.unescape(text_in)

    # 2. Xóa các đề cập người dùng: @username
    text_in = re.sub(r'@[^\s]+', '', text_in)

    # 3. Chỉ xóa ký hiệu hashtag (#tag → tag)
    text_in = re.sub(r'#(\w+)', r'\1', text_in)

    # 4. Xóa các đường dẫn URL
    text_in = re.sub(r'https?://\S+|www\.\S+', '', text_in)

    # 5. Xóa emoji và các ký hiệu không phải văn bản
    text_in = re.sub(
        r'[\U00010000-\U0010ffff]',
        '',
        text_in
    )

    # 6. Xóa các ký tự đặc biệt (giữ lại chữ cái, số và dấu tiếng Việt)
    text_in = re.sub(r'[^\w\s]', ' ', text_in)

    # 7. Chuẩn hóa khoảng trắng (nhiều khoảng trắng → một khoảng trắng)
    text_in = re.sub(r'\s+', ' ', text_in).strip()

    return text_in
```

```python
merged_df['content_text'] = merged_df['content'].apply(clean_text) # Xem trước một số văn bản đã làm sạch
merged_df.head()
```

```python
# Lặp lại nhiều lần với bước sau
import pandas as pd
from collections import Counter
import re

def stats_short_words_df(df, text_col, min_len=1, max_len=2, top_k=50):
    """
    Thống kê các từ có độ dài từ min_len đến max_len.
    """
    counter = Counter()

    for text in df[text_col]:
        # Kiểm tra dữ liệu rỗng hoặc không phải chuỗi
        if pd.isna(text) or not isinstance(text, str):
            continue

        # 1. Chuyển về chữ thường
        text = text.lower()

        # 2. Dùng Regex để chỉ lấy chữ cái và số (loại bỏ dấu câu)
        # \w+ nghĩa là lấy các cụm từ liền mạch
        words = re.findall(r'\w+', text)

        # 3. Lọc độ dài từ
        short_words = [w for w in words if min_len <= len(w) <= max_len]

        # 4. Cập nhật vào bộ đếm
        counter.update(short_words)

    return counter.most_common(top_k)

# --- CÁCH SỬ DỤNG ---
# Giả sử bạn có DataFrame là merged_df
top_words = stats_short_words_df(merged_df, "content_text", top_k=2000)

print("--- CÁC TỪ NGẮN XUẤT HIỆN NHIỀU NHẤT ---")
for word, freq in top_words:
    print(f"Từ '{word}': xuất hiện {freq} lần")
```

```python
# Lặp lại nhiều lần để chuẩn hóa
import re

# 1. DANH SÁCH TỪ ĐIỂN (Key phải viết thường toàn bộ)
NORMALIZE_MAP = {
    "k": "không",
    "ko": "không",
    "kh": "không",
    "kg": "không",
    "dc": "được",
    "đc": "được",
    "sp": "sản_phẩm",
    "ng": "người",
    "ok": "ổn",
    "ad": "quản_trị_viên",
    "ib": "nhắn_tin",
    "rep": "trả_lời",
    "shop": "cửa_hàng",
    "mn": "mọi_người",
    "vs": "với",
    "mk": "mình",
    "r": "rồi",
    "khong": "không",
    "km": "Khuyến_mãi",
    "oke": "ổn",
    "oki": "ổn",
    "okê": "ổn",
    "ms": "mới",
    "cx": "cũng",
    "of": "của",
    "or": "hoặc",
    "sdt": "số_điện_thoại"

    # Bạn bổ sung thêm từ dựa trên kết quả của Phần 1
}

def normalize_text(text, norm_map):
    """
    Chuẩn hóa văn bản dựa trên từ điển map.
    """
    if not isinstance(text, str):
        return "" # Trả về chuỗi rỗng nếu dữ liệu lỗi

    # Bước 1: Chuyển toàn bộ về chữ thường
    text = text.lower()

    # Bước 2: Thay thế dấu câu bằng khoảng trắng để tách từ dính dấu
    # Ví dụ: "hàng ok,đẹp" -> "hàng ok đẹp"
    text = re.sub(r'[^\w\s]', ' ', text)

    # Bước 3: Tách từ ra thành list
    words = text.split()

    # Bước 4: So khớp và thay thế từ (Map lookup)
    # Nếu từ có trong map thì thay, không thì giữ nguyên
    new_words = [norm_map.get(w, w) for w in words]

    # Bước 5: Ghép lại thành câu
    return " ".join(new_words)

# --- CÁCH SỬ DỤNG ---
# Áp dụng hàm này vào cột dữ liệu
merged_df['content_text'] = (
    merged_df['content_text']
    .apply(lambda x: normalize_text(x, NORMALIZE_MAP))
)

# Kiểm tra kết quả
print("--- KẾT QUẢ SAU KHI CHUẨN HÓA ---")
merged_df[["content_text"]].head()
```

```python
def load_vietnamese_stopwords(path: str) -> set:
    """
    Load Vietnamese stopwords from .txt file
    Each stopword is on a separate line
    """
    with open(path, encoding='utf-8') as f:
        stopwords = {line.strip() for line in f if line.strip()}
    return stopwords
```

```python
def remove_vn_stopwords(text: str, stopword_set: set) -> str:
    """
    Remove Vietnamese stopwords from text
    - No tokenization library
    - No stemming
    - No word-length filtering
    """
    words = text.split()
    words = [w for w in words if w not in stopword_set]
    return " ".join(words)
```

```python
# Tải stopwords (chỉ thực hiện một lần)
vn_stopwords = load_vietnamese_stopwords(
    root_path + "vietnamese-stopwords.txt"
)

# Làm sạch văn bản
merged_df["content_text"] = merged_df["content_text"].apply(
    lambda x: remove_vn_stopwords(x, vn_stopwords)
)

# Xem trước một số văn bản đã làm sạch
merged_df[["content","content_text"]].head(10000).reset_index(drop=True)  # <--- khuyến khích xem nhiều lần để chuẩn hóa được dữ liệu
```

#5. Tạo các đặc trưng dựa theo các tiêu chí để chuẩn bị gán nhãn

##5.1. Hành vi đánh giá ồ ạt

```python
# sắp xếp theo user và thời gian
merged_df = merged_df.sort_values(['customer_id', 'created_at'])

# tính khoảng cách giữa 2 comment liên tiếp của cùng user
merged_df['delta_minutes'] = (
    merged_df
        .groupby('customer_id')['created_at']
        .diff()
        .dt.total_seconds() / 60
)

valid_delta = merged_df.loc[
    merged_df['delta_minutes'] > 0,
    'delta_minutes'
]

threshold_burst = valid_delta.quantile(0.05)

print("Ngưỡng burst (phút):", threshold_burst)
```

```python
# lấy ngưỡng khoảng cách bất thường giữa 2 cmt là 5% khoảng cách ngắn nhất giữa 2 lần cmt đã tổng hợp ở trên
threshold_burst = merged_df['delta_minutes'].quantile(0.05)
print("Ngưỡng burst (phút):", threshold_burst)
```

```python
# Xác định những khoảng nhỏ là những dấu hiệu về một bình luận bất thường
merged_df['small_gap'] = merged_df['delta_minutes'] <= threshold_burst
```

```python
# Góm nhóm những lần bất thường của các bình luận và đếm
merged_df['burst_count'] = (
    merged_df.groupby('customer_id')['small_gap']
    .transform(lambda x: x * (x.groupby((x != x.shift()).cumsum()).cumcount() + 1))
)
```

```python
# Xác định và nghi ngờ những bình luận đó có phải là gian lận hay không bằng cách đếm xem nếu có lơn hơn hoặc bằng 4 lần diễn ra các khoảng nhỏ diễn ra liên tiếp nhau
merged_df['F3_flag'] = merged_df['burst_count'] >= 4
```

```python
# 1. Lấy danh sách user đã từng bị gắn cờ F3
fraud_users = merged_df.loc[
    merged_df['F3_flag'] == True,
    'customer_id'
].unique()

# 2. Lọc toàn bộ comment của những user đó
result = (
    merged_df[merged_df['customer_id'].isin(fraud_users)]
    [['customer_id',
      'created_at',
      'purchased_at',
      'delta_minutes',
      'burst_count',
      'F3_flag']]
    .sort_values(['customer_id', 'created_at'])
)

# 3. In 15 dòng đầu
# Kết quả nó sẽ hiện những lần cmt của các user đã bị gắn cờ gian lận và hiện từ lúc mà burst count từ 0 đến 4
print(result.head(50))
```

##5.2. Sự tương đồng về nội dung

```python
# Gom nhóm comment theo customer_id
user_comments = merged_df.groupby("customer_id", sort=False)["content"].apply(list).reset_index()

# In 5 dòng đầu của bảng sau khi gom
print(user_comments.head())

# Xem chi tiết comment của customer id đầu tiên
print("\nCustomer ID:", user_comments.loc[0, "customer_id"])
print("Danh sách comment:")

for cmt in user_comments.loc[0, "content"]:
    print("-", cmt)
```

```python
# Tạo cột và tính giá trị
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# tạo 2 cột mới
user_comments["mean_similarity"] = np.nan
user_comments["max_similarity"] = np.nan
vectorizer = TfidfVectorizer()

#tính mean,max
for i in range(len(user_comments)):
    comments = user_comments.loc[i, "content"]
    # loại comment rỗng
    comments = [c for c in comments if isinstance(c, str) and c.strip() != ""]
    # nếu user có <2 comment thì bỏ qua
    if len(comments) < 2:
        continue
    try:
        tfidf = vectorizer.fit_transform(comments)
        sim_matrix = cosine_similarity(tfidf)
        sim_values = sim_matrix[np.triu_indices_from(sim_matrix, k=1)]
        user_comments.loc[i, "mean_similarity"] = sim_values.mean()
        user_comments.loc[i, "max_similarity"] = sim_values.max()
    except:
        continue
# in kết quả
print(user_comments.head())
```

```python
#Xác định ngưỡng
import numpy as np

# Bỏ các user không tính được similarity
valid_users = user_comments.dropna(subset=["mean_similarity", "max_similarity"])

# Tính mean,max và std của toàn dataset
mean_mean = valid_users["mean_similarity"].mean()
std_mean = valid_users["mean_similarity"].std()

mean_max = valid_users["max_similarity"].mean()
std_max = valid_users["max_similarity"].std()

# Tính ngưỡng
threshold_mean = mean_mean + std_mean
threshold_max = mean_max + std_max

# In kết quả
print("\nNgưỡng phát hiện spam:")
print("Threshold mean =", threshold_mean)
print("Threshold max =", threshold_max)
```

```python
# Gán nhãn
user_comments["F4_Flag"] = (
    (user_comments["mean_similarity"] >= threshold_mean) |
    (user_comments["max_similarity"] >= threshold_max)
)

# in 15 dòng đầu và hiển thị tất cả cột
print(user_comments.head(15))
```

```python
merged_df = merged_df.merge(
    user_comments[["customer_id", "mean_similarity", "max_similarity", "F4_Flag"]],
    on="customer_id",
    how="left"
)

print(merged_df.head())
```

## 5.3. Bất thường về độ trễ thời gian

```python
# Tính khoảng cách thời gian giữa mua và đánh giá (phút)
merged_df['time_gap_minutes'] = (
    (merged_df['created_at'] - merged_df['purchased_at'])
    .dt.total_seconds() / 60
)

# Loại bỏ các giá trị âm (nếu có lỗi data)
valid_gap = merged_df.loc[
    merged_df['time_gap_minutes'] > 0,
    'time_gap_minutes'
]
```

```python
# Xác định ngưỡng bất thường (5% nhỏ nhất)
threshold_timegap = valid_gap.quantile(0.05)

print("Ngưỡng time-gap bất thường (phút):", threshold_timegap)

#  Gắn cờ F5
merged_df['F5_flag'] = (
    merged_df['time_gap_minutes'] > 0
) & (
    merged_df['time_gap_minutes'] <= threshold_timegap
)
```

```python
# Lấy danh sách user đã từng bị gắn cờ F5
fraud_users_F5 = merged_df.loc[
    merged_df['F5_flag'] == True,
    'customer_id'
].unique()

# Lọc toàn bộ comment của những user đó
result_F5 = (
    merged_df[merged_df['customer_id'].isin(fraud_users_F5)]
    [['customer_id',
      'created_at',
      'purchased_at',
      'time_gap_minutes',
      'F5_flag']]
    .sort_values(['customer_id', 'created_at'])
)

#  In 50 dòng đầu
print(result_F5.head(50))
```

#6. Chọn ra 10% dữ liệu

```python
# Lọc spam và ham theo đúng logic
spam_df = merged_df[
    (merged_df["F3_flag"] == True) |
    (merged_df["F4_Flag"] == True) |
    (merged_df["F5_flag"] == True)
]

ham_df = merged_df[
    (merged_df["F3_flag"] == False) &
    (merged_df["F4_Flag"] == False) &
    (merged_df["F5_flag"] == False)
]

# Tính số lượng cần lấy cho 10% tổng dữ liệu
sample_size = int(len(merged_df) * 0.1)
sample_per_class = sample_size // 2  # Chia đều cho spam và ham

# Lấy mẫu cân bằng
spam_sample = spam_df.sample(n=min(sample_per_class, len(spam_df)), random_state=42)
ham_sample = ham_df.sample(n=min(sample_per_class, len(ham_df)), random_state=42)

# Kết hợp và xáo trộn
final_df = pd.concat([spam_sample, ham_sample]).sample(frac=1, random_state=42)

path = "/content/drive/MyDrive/big_data/phan_1/data/10%_dataset.csv"
final_df.to_csv(path, index=False)
```
